**Reasoning**:
Display a few sample images and their corresponding annotations to visualize the data and verify the annotation format.



```python
import matplotlib.pyplot as plt
import cv2

def plot_image_with_annotations(image_path, label_path, classes):
    """Reads and displays an image with YOLO format annotations."""
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, _ = img.shape

    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    ax = plt.gca()

    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            for line in f.readlines():
                class_id, center_x, center_y, width, height = map(float, line.strip().split())

                # Convert YOLO format to bounding box coordinates
                x_min = int((center_x - width / 2) * w)
                y_min = int((center_y - height / 2) * h)
                box_width = int(width * w)
                box_height = int(height * h)

                rect = plt.Rectangle((x_min, y_min), box_width, box_height,
                                     fill=False, color='red', linewidth=2)
                ax.add_patch(rect)
                label = classes[int(class_id)]
                ax.text(x_min, y_min - 10, label, color='red', fontsize=12)
    else:
        print(f"No label file found for {os.path.basename(image_path)}")


    ax.set_title(os.path.basename(image_path))
    ax.axis('off')
    plt.show()

extraction_dir = '/content/dataset'
train_images_dir = os.path.join(extraction_dir, data_config.get('train', ''))
train_labels_dir = train_images_dir.replace('images', 'labels')
classes = data_config.get('names', {})

# Display a few training images and annotations
print("\nDisplaying sample training images with annotations:")
sample_images = os.listdir(train_images_dir)[:3] # Get first 3 image filenames

for img_file in sample_images:
    image_path = os.path.join(train_images_dir, img_file)
    label_path = os.path.join(train_labels_dir, os.path.splitext(img_file)[0] + '.txt')
    plot_image_with_annotations(image_path, label_path, classes)
```

**Reasoning**:
The error indicates that the annotation format is not the standard YOLO format (class_id center_x center_y width height). It appears there are more than 5 values per line in the label files, which suggests it might be OBB (Oriented Bounding Box) format. I need to adjust the parsing logic in `plot_image_with_annotations` to handle OBB format which typically includes 8 coordinates (x1, y1, x2, y2, x3, y3, x4, y4) plus the class id. I will modify the plotting function to read 9 values per line (class_id and 8 coordinates) and draw the OBB polygon.



```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os

def plot_image_with_annotations(image_path, label_path, classes):
    """Reads and displays an image with YOLO OBB format annotations."""
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, _ = img.shape

    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    ax = plt.gca()

    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            for line in f.readlines():
                values = list(map(float, line.strip().split()))
                if len(values) == 9: # Assuming OBB format: class_id x1 y1 x2 y2 x3 y3 x4 y4
                    class_id = int(values[0])
                    coords = np.array(values[1:]).reshape(-1, 2) * np.array([w, h])
                    polygon = coords.astype(int)

                    # Draw the polygon
                    ax.add_patch(plt.Polygon(polygon, fill=False, color='red', linewidth=2))

                    # Place label near one of the points (e.g., top-left)
                    label = classes.get(class_id, f'Unknown Class {class_id}')
                    # Find the top-left most point to place the label
                    min_x, min_y = np.min(polygon, axis=0)
                    ax.text(min_x, min_y - 10, label, color='red', fontsize=12)
                else:
                    print(f"Skipping line in {label_path} due to incorrect number of values: {len(values)}")

    else:
        print(f"No label file found for {os.path.basename(image_path)}")

    ax.set_title(os.path.basename(image_path))
    ax.axis('off')
    plt.show()

extraction_dir = '/content/dataset'
data_yaml_path = '/content/dataset/data.yaml'
with open(data_yaml_path, 'r') as f:
    data_config = yaml.safe_load(f)

train_images_dir = os.path.join(extraction_dir, data_config.get('train', ''))
train_labels_dir = train_images_dir.replace('images', 'labels')
classes = data_config.get('names', {})

# Display a few training images and annotations
print("\nDisplaying sample training images with annotations (OBB format):")
sample_images = os.listdir(train_images_dir)[:3] # Get first 3 image filenames

for img_file in sample_images:
    image_path = os.path.join(train_images_dir, img_file)
    label_path = os.path.join(train_labels_dir, os.path.splitext(img_file)[0] + '.txt')
    plot_image_with_annotations(image_path, label_path, classes)
```

## Define a hyperparameter grid

### Subtask:
Define a hyperparameter grid by specifying a range of values for hyperparameters like learning rate, epochs, and batch size to experiment with.


**Reasoning**:
Define a hyperparameter grid by specifying a range of values for hyperparameters like learning rate, epochs, and batch size to experiment with.



```python
param_grid = {
    'lr0': [0.01, 0.001],
    'epochs': [50, 100],
    'batch': [16, 32],
}

print("Hyperparameter grid defined:")
print(param_grid)
```

## Train models with different hyperparameters

### Subtask:
Train models with different hyperparameters.


**Reasoning**:
Iterate through the hyperparameter grid and train the model for each combination, storing the results.



```python
from ultralytics import YOLO
import itertools
import os

best_mAP50 = 0.0
best_params = None
all_results = {}

data_yaml_path = '/content/dataset/data.yaml'

# Iterate through all combinations of hyperparameters
keys, values = zip(*param_grid.items())
for v in itertools.product(*values):
    current_params = dict(zip(keys, v))
    lr0 = current_params['lr0']
    epochs = current_params['epochs']
    batch = current_params['batch']

    print(f"\nTraining with params: lr0={lr0}, epochs={epochs}, batch={batch}")

    # Load a fresh pre-trained model for each run
    model = YOLO('yolov8s.pt')

    # Define a unique name for the training run directory
    run_name = f'train_lr{lr0}_ep{epochs}_bs{batch}'

    # Train the model
    results = model.train(
        data=data_yaml_path,
        epochs=epochs,
        imgsz=640,
        lr0=lr0,
        batch=batch,
        name=run_name # Save results to a unique directory
    )

    # Store the training results and evaluation metrics
    all_results[run_name] = {
        'params': current_params,
        'metrics': results.results_dict # Access the validation metrics from the training results
    }

    # Track the best performing model based on mAP50
    current_mAP50 = results.results_dict.get('metrics/mAP50(B)', 0.0)
    if current_mAP50 > best_mAP50:
        best_mAP50 = current_mAP50
        best_params = current_params

print("\nHyperparameter tuning complete.")
print(f"Best mAP50: {best_mAP50}")
print(f"Best parameters: {best_params}")

# You can now analyze the all_results dictionary to compare different runs
```
